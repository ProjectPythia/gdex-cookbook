## API Conventions

All endpoints live under `https://gdex.ucar.edu/api/`. Dataset IDs use the format `d######` (six digits). Every successful response follows this envelope:

```json
{
  "status": "ok",
  "http_response": 200,
  "error_messages": [],
  "contact": "datahelp@ucar.edu",
  "data": { ... }
}
```

## Available Data Services

TDS provides multiple access protocols for each dataset. Common ones include:
- **OPeNDAP** — remote array subsetting, works with xarray
- **NCSS** — NetCDF Subset Service, time/spatial subsetting
- **WMS** — Web Map Service for visualization

## Overview

GDEX provides Analysis-Ready Cloud-Optimized (ARCO) versions of select datasets in Zarr or virtualized Zarr (kerchunk, virtualizarr) format. ARCO data is chunked and indexed for fast/parallel access — ideal for large-scale analysis like machine learning or AI workflow.

1. Check whether a dataset has ARCO data available
2. Browse and search ARCO variables
3. Obtain the URL link to the ARCO data

## Find ARCO data on GDEX

To find ARCO data on GDEX, use the dedicated search filter:

1. Navigate to the [GDEX web portal](https://gdex.ucar.edu)
2. Press the **AI Ready Datasets** button under the search bar to display all available Analysis-Ready Cloud-Optimized datasets
3. One can also press the **Zarr Format Datasets** button for the all the available zarr format dataset.

:::{figure} ../../images/gdex-arco.png
:alt: GDEX ARCO filter button under the search bar
:width: 90%

Click the **AI Ready Datasets** or **Zarr Format Datasets** button to show ARCO datasets available in Zarr or virtualized Zarr formats
:::


## Access the ARCO data

Once you open a dataset page, use the **Data Access** tab at the top to view available data links. In the column view, scroll to the desired row (or use the all-data link).

:::{figure} ../../images/gdex-zarr-dataaccess.png
:alt: GDEX Zarr data access landing page
:width: 90%

Select the data group you want to explore.
:::

The selected link opens a subgroup view of the dataset. Choose the desired subgroup (for example, a specific time frequency).

:::{figure} ../../images/gdex-zarr-subgroup.png
:alt: GDEX Zarr data subgroup view
:width: 90%

Select the subgroup you want to use.
:::

You will then reach the final data file page, where you can click the **Copy URL** button to get the Zarr store access URL (example shown: CESM2 Large Ensemble in Zarr format).

:::{figure} ../../images/gdex-zarr-access.png
:alt: GDEX Zarr data URL link
:width: 90%

Click the Zarr store **Copy URL** button to copy the access link.
:::
